In the next code, we have chosen the columns and kept only those which will be required for testing.

```python
training_set = dataset_train.iloc[:, 1:2].values
```

## Checking for autocorrelation

```python
from pandas.plotting import autocorrelation_plot
autocorrelation_plot(training_set)
plt.show()
```

From the output of the graph we can see that as the graph is well beyond the confidence interval, there exists auto correlation in the dataset. This also means that the dataset is not stationary and at least a first level of differencing will be required.

# Fitting the ARIMA Model

To fit the ARIMA model, we need to import ARIMA from statsmodels library of python. We have also imported the dataframe library so that we could collect all the residuals and make a plot out of them.

ARIMA has three components:
i. Autoregression - This component represents the lag present in the dataset, denoted by p
ii. Intergrated - This represents the number of times the dataset had to be differenced. Differencing is required because the other parts of the model can only work on stationary dataset. By differencing we aim to get the key statistical components, that is mean and variance of the set constant. It is denoted y d
iii. Moving Average - It represents the relation of the model with its previous residual error. It is denoted by q.
Thus, an ARIMA model can be represented finally as ARIMA(p,d,q).

```python
from pandas import datetime
from pandas import DataFrame
from statsmodels.tsa.arima_model import ARIMA
# fit model
model = ARIMA(training_set, order=(2,1,2))
model_fit = model.fit(disp=0)
print(model_fit.summary())
# plot residual errors
residuals = DataFrame(model_fit.resid)
residuals.plot()
plt.show()
residuals.plot(kind='kde')
plt.show()
print(residuals.describe())
```

## ARIMA Output

The ARIMA that best fits the model is of the order (2,1,2). We have tried with different lag and moving average levels. The AIC (Akaike Information Criterion) was observed to be the best in this case. It maybe noted that removing the moving average component would not have affected the output of this model much. However, the final prediction was more accurate with it.

Further, the graphs following the statistical output need some explaination. The first graph is the residuals. It can be seen that it represents a white noise. The graph following it also shows a normal distribution of residuals. Thus, it can be concluded from thes two as well that the ARIMA model with the order (2,1,2) fit it well.

# Predicting the Output

In the following code, we have input the test data which we already created.

```python
dataset_test = pd.read_excel('/content/drive/My Drive/Nifty_Test.xlsx')
real_stock_price = dataset_test.iloc[:, 1:2].values
```

In the following set of code, we have created a function which calls the ARIMA model to fit and forecast the data.

```python
#Function that calls ARIMA model to fit and forecast the data
def StartARIMAForecasting(Actual, P, D, Q):
	model = ARIMA(Actual, order=(P, D, Q))
	model_fit = model.fit(disp=0)
	prediction = model_fit.forecast()[0]
	return prediction
```

```python
#new arrays to store actual and predictions
Actual = [x for x in training_set]
Predictions = list()
```

In the following code, we have created a for loop which predicts the values using ARIMA and adds them to a lsit called prediction. This list will later be utilised to determine the performance metrics and visualize the results.

```python
#in a for loop, predict values using ARIMA model
for timepoint in range(len(real_stock_price)):
	ActualValue =  real_stock_price[timepoint]
	#forcast value
	Prediction = StartARIMAForecasting(Actual, 2,1,2)    
	#add it in the list
	Predictions.append(Prediction)
	Actual.append(ActualValue)
```

## Visualization and Performance metrics

```python
plt.plot(real_stock_price, color = 'red', label = 'Real Nifty Price')
plt.plot(Predictions, color = 'blue', label = 'Predicted Nifty Price')
plt.title('Nifty Price Prediction')
plt.xlabel('Time')
plt.ylabel('Nifty Price')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_squared_log_error

print("The mean squared error (MSE) = ", mean_squared_error(real_stock_price, Predictions))
print("The mean absolute error (MAS) = ", mean_absolute_error(real_stock_price, Predictions))
print("The mean squared log error (MSLE) = ", mean_squared_log_error(real_stock_price, Predictions))
```
